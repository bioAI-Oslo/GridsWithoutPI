```python
import torch
import numpy as np 
import scipy

import matplotlib.pyplot as plt
import spatial_maps as sm

plt.style.use("project_style.mplstyle")

from tqdm import tqdm
from models import FFGC, RNNGC
from dataset import DatasetMaker
from plotting_functions import *
```

```python
def shuffle_color(n):
    for i in range(n):
        plt.plot(0)
```

```python
gs_lims = None
o_lims = (np.array([0, np.pi/3]) -0.05 )  *1.1
s_lims = [0, 2*np.pi]

# 1D statistics
for statistic, lims in zip(("Grid Score", "Orientation", "Spacing"), [gs_lims, o_lims, s_lims]) :
    for model in ("RNN", "FF"):
        stat = np.load(f"./statistics/{model}_{statistic}.npy")
        shuffle_color(4)

        plt.hist(stat, label = model, alpha = 0.5, bins = 16, density=False)
        plt.title(statistic)
        plt.xlim(lims)
        shuffle_color(0)
    plt.legend()
    plt.savefig(f"./figures/{model}_{statistic}.png", dpi = 300)
    plt.show()
```

```python
FF_phases = np.load("./statistics/FF_phases.npy")
RNN_phases = np.load("./statistics/RNN_phases.npy")
plt.figure(figsize=(2,1.5))
shuffle_color(2)
plt.plot(RNN_phases[:,0], RNN_phases[:,1],"o", alpha = 0.5, label = "RNN", markersize = 2, markeredgecolor = "none")

#plt.axis()
shuffle_color(1)
plt.plot(FF_phases[:,0], FF_phases[:,1],"o", alpha = 0.5, label = "FF", markersize = 2, markeredgecolor = "none")
plt.legend()
# plt.axis([-10, 70, -10, 70])
plt.axis("square")
plt.savefig("./figures/phases.png", dpi = 300)
```
